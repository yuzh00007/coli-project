```python
import pandas as pd
```

```python
def read_json_file(file_path):
    try:
        json_obj = pd.read_json(path_or_buf=file_path, lines=True, orient=str)
    except FileNotFoundError as error:
        raise FileNotFoundError(
            f"{error}"
            f"could not read file - make sure the path is correct."
            f"current working dir: {os.getcwd()}"
        )

    return json_obj
```

```python
def generate_results_table(file_name):
    data = read_json_file(file_name)

    final = data[data.loss.isnull() & data.train_loss.isnull()]
    
    for col in [
        "loss", "grad_norm", "learning_rate", "train_runtime", 
        "train_samples_per_second", "train_steps_per_second", "train_loss",
        "eval_samples_per_second", "eval_steps_per_second", "eval_runtime"
    ]:
        final = final.drop(col, axis=1)

    return final
```
